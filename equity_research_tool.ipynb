### It has problems of spltting text based on different splitters, as some chunks are more than size, (\n , . , ? )

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator='\n',
    chunk_size=200,
    chunk_overlap=0
)

chunks = splitter.split_text(text)
len(chunks)
```

```python
for chunk in chunks:
    print(len(chunk))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    # U may wonder why we added 'space' as this may break down every word, it would first split and then merge for optimization
    separators=['\n\n', '\n', '.', '?', ',', ' '],
    chunk_size=200,
    chunk_overlap=0
)

chunks_r = r_splitter.split_text(text)
```

```python
for chunk in chunks_r:
    print(len(chunk))

len(chunks_r)
```
